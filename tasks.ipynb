```python
#Task 5

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame for easier manipulation
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map the target labels to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

# Choose two features from the dataset for the scatter plot
feature_1 = 'sepal length (cm)'
feature_2 = 'petal length (cm)'

# Create a scatter plot with color-coded classes
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=feature_1, y=feature_2, hue='species', palette='Set1', s=100)

# Add title and labels
plt.title(f"Scatter Plot of {feature_1} vs {feature_2}", fontsize=16)
plt.xlabel(feature_1, fontsize=12)
plt.ylabel(feature_2, fontsize=12)

# Display the plot
plt.show()
```

```python
# Task 6

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris

# Load the iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame for easier manipulation
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map the target labels to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

# Choose two features from the dataset for the scatter plot
feature_1 = 'sepal length (cm)'
feature_2 = 'petal length (cm)'

# Create the scatter plot with color-coded classes
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x=feature_1, y=feature_2, hue='species', palette='Set1', s=100)

# Fit a linear regression line using numpy.polyfit
x_values = df[feature_1]
y_values = df[feature_2]

# Fit a line of degree 1 (linear regression)
slope, intercept = np.polyfit(x_values, y_values, 1)

# Generate the regression line values
regression_line = slope * x_values + intercept

# Plot the regression line
plt.plot(x_values, regression_line, color='black', linestyle='--', label='Regression Line')

# Add title and labels
plt.title(f"Scatter Plot with Regression Line of {feature_1} vs {feature_2}", fontsize=16)
plt.xlabel(feature_1, fontsize=12)
plt.ylabel(feature_2, fontsize=12)

# Show the legend
plt.legend()

# Display the plot
plt.show()

```

```python
# Task 7

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Convert to a Pandas DataFrame
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target

# Map the numeric target to species names
species_map = {0: 'setosa', 1: 'versicolor', 2: 'virginica'}
df['species'] = df['species'].map(species_map)

# Create a box plot with palette and updated parameters
plt.figure(figsize=(8, 6))
sns.boxplot(
    x='species',
    y='petal length (cm)',
    hue='species',               # Needed for custom palette in future versions. When I initially ran the code, I got a warning, so I had to change it accordingly to prevent this in the future.
    data=df,
    palette='Set2',
    dodge=False,                 # Ensures boxes aren't duplicated for hue
    legend=False                 # We don't need a redundant legend since x-axis shows species
)

# Add labels and title
plt.title("Box Plot of Petal Length by Iris Species", fontsize=16)
plt.xlabel("Species", fontsize=12)
plt.ylabel("Petal Length (cm)", fontsize=12)

# Show the plot
plt.show()

# References

# Waskom, M., Botvinnik, O., O'Kane, D., Hobson, P., Lukauskas, S., Gemperline, D., … & Qalieh, A. (2020). Seaborn: v0.11.0 (July 2020). *Zenodo*. https://doi.org/10.5281/zenodo.3961452
# Hunter, J. D. (2007). Matplotlib: A 2D graphics environment. *Computing in Science & Engineering, 9*(3), 90-95. https://doi.org/10.1109/MCSE.2007.55
# McKinney, W. (2018). *Python for Data Analysis: Data Wrangling with Pandas, NumPy, and Jupyter (2nd ed.)*. O'Reilly Media.
# Pedregosa, F., Varoquaux, G., Gramfort, A., Michel, V., Thirion, B., Grisel, O., … & Duchesnay, E. (2011). Scikit-learn: Machine learning in Python. *Journal of Machine Learning Research, 12*, 2825-2830. https://scikit-learn.org/
# Fisher, R. A. (1936). The use of multiple measurements in taxonomic problems. *Annals of Eugenics, 7*(2), 179–188. https://doi.org/10.1111/j.1469-1809.1936.tb02137.x

# Additional Reference for Seaborn Box Plot Update

# Waskom, M. (2023). *Seaborn v0.13.0 Release Notes*. Retrieved from https://seaborn.pydata.org/whatsnew/v0.13.0.html
# Waskom, M. et al. (n.d.). *Seaborn Documentation: Categorical Plotting Functions*. Retrieved from https://seaborn.pydata.org/generated/seaborn.boxplot.html
```
